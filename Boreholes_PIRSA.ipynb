```python
# lets perform some transformatio to plot the location of those bore holes
df["shape"][1]
# shape has the lon and lat information but lets split it into 2 columns so we can put then 
#in a array to be used by folium API
```

```python
df["shape"] = df["shape"].str.replace(")","")
df["shape"] = df["shape"].str.replace("(","")
df["lon"] = df["shape"].apply(lambda x: x.split()[1])
df["lat"] = df["shape"].apply(lambda x: x.split()[2])
df["lon"] = df["lon"].astype(float)
df["lat"] = df["lat"].astype(float)
df[['lat','lon']].values[1]
```

```python
# plot location minearl boreholes test

point = df[['lat','lon']].values[1]
import folium
# plot 
m = folium.Map(location=[-30.85179213, 132.1434783 ],
               zoom_start=6, 
               tiles='Stamen Terrain')
folium.Circle(
    radius=100,
    location=point,
    color='blue',
    fill=True,
    ).add_to(m)

m
```

The folium api is excelent and really looks nice and it is interactive however it does not work well with too many data poits. Lets slice the dataset in few by type and plot it in different colors to have an idea of where there were dug in SA.

will break into 4 categories

- Mineral
- Petroleum
- Engineering
- Water

Will only pick a sample (random) of 500 holes for water as there are too many. I believe will give a goos representation of where they are in SA. I am more interested in where the other holes are located in the state 

```python
water = df[df['description']=='water borehole'].sample(500,random_state=0)



# plot location water boreholes 500 random picked

m = folium.Map(location=[-30.85179213, 132.1434783 ],
               zoom_start=5, 
               tiles='Stamen Terrain')
for point in water[['lat','lon']].values:
    folium.Circle(
        radius=100,
        location=point,
        color='blue',
        fill=True,
        ).add_to(m)

m

```

```python
# plot diferent color for different description

descriptions = ['water borehole','petroleum borehole','mineral borehole','engineering borehole']
colors = ['blue','red','green', 'purple']

#copy random sample of water
data = water.copy()

for i in descriptions[1:]:
    data = data.append(df[df['description']==i])
    
m = folium.Map(location=[-30.85179213, 132.1434783 ],
               zoom_start=5, 
               tiles='Stamen Terrain')
for description, color in zip(descriptions,colors):
    for point in data[data['description']==description][['lat','lon']].values:
        folium.Circle(
            radius=100,
            location=point,
            color=color,
            fill=True,
            ).add_to(m)

# save html 
m.save('SA_boreholes.html')        

#display map
m


    
```

```python
df['drillEndDate'] = pd.to_datetime(df.drillEndDate)
df['year'] = df['drillEndDate'].apply(lambda x: x.year)
```

```python
df.year.value_counts().sort_index().plot(figsize=(15,10),title='all holes by year');
```

```python
df.year.value_counts()
```

```python
descriptions = ['water borehole','petroleum borehole','mineral borehole','engineering borehole']
colors = ['blue','red','green', 'purple']

fig, ax = plt.subplots(figsize=(15,10))

for description, color in zip(descriptions,colors):
    dt = df[df['description']==description]
    summary = dt.year.value_counts()
    plt.bar(summary.index,summary.values,label=description)
    
    

plt.legend();
```

Where the top 10 deepest holes are located in SA



```python
deepest = df.sort_values(by='boreholeLength_m',ascending=False)[:10]

m = folium.Map(location=[-30.85179213, 132.1434783 ],
               zoom_start=5, 
               tiles='Stamen Terrain')

i=0
for _, row in deepest[['lat','lon','description','boreholeLength_m']].iterrows():
    point = [row['lat'],row['lon']]
    des = row['description']
    depth = row['boreholeLength_m']
    i += 1
    folium.Marker(location=point
                 ,tooltip=f'depth rank {i}'
                 ,popup=f'type: {des}').add_to(m)

    

#display map
m

```
